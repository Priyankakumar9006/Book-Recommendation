```python
data_cleaned2=data_cleaned1.drop_duplicates('userid')
data_cleaned2.shape
```

```python
user_df = data_cleaned2.pivot(index='userid',columns='booktitle',values='bookrating').reset_index(drop=True)
```

```python
user_df
```

```python
user_df.index =data_cleaned2.userid.unique()
```

```python
user_df
```

```python
#Imputing NaNs with 0 values
user_df.fillna(0, inplace=True)
```

```python
user_df
```

```python
#Calculating Cosine
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine, correlation
```

```python
user_sim = 1 - pairwise_distances( user_df.values,metric='cosine')
```

```python
user_sim
```

```python
#Store the results in a dataframe
user_sim_df = pd.DataFrame(user_sim)
```

```python
user_sim_df.index = data_cleaned2.userid.unique()
user_sim_df.columns = data_cleaned2.userid.unique()
```
